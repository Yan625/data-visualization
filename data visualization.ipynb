# Data visualization

    1. data visualization refers to techniques used to communicate insights from data through visual            
       representation.
    
    2. it means, presenting data in a visual manner, using charts,graphs and maps.
    
    3. data visualization is an art of presenting the data in a manner that any non-technical person can 
       understand it
    
    4. it allows data seicntists to converse with their end users
    
    5. need of data visualization:
      
       5.1 to analyze massive amount of data and make decisions based on data
       
       5.2 by using visual elements like charts, graphs and maps, data visualization tools provide and accessible 
           way to see and understand trends, outliers and patterns in data
       
       5.3 the amount of information that humans again through vision is far beyond that of other organs
       
       5.4 visualization can help us deal with more complex information and enhance memory
       
    6. importance of data visualization:
       
       6.1 help in quichly understanding the data
       
       6.2 simplifies large datasets in a way that people can easily interpret
       
       6.3 shows trends that may not be apprent in first glance
       
       6.4 identify areas that need attention or improvment
       
       6.5 offers new insights to your data
       
       6.6 quickly identify ourliers
     
    7. types of data visualization
       
       7.1 exploratory analysis: find insights
                
                7.1.1 it is done during data analysi to find insights
                
                7.1.2 this is aappropriate when you have whole bunch of data and you are not sure what's in it
                
                7.1.3 visualization built for these purposes do not need to be perfect. you are simply looking 
                      for patterns
                     
       7.2 explanatory analysis: present insights
            
                7.2.1 this is done after you find a ninsight
                
                7.2.2 this is appropriate when you already know what the data has to say, and you are tring to 
                      tell that story to somebody else
                
                7.2.3 visualizatino done for these purposes need to be surrounded by a story that points the 
                      reader to an insight that answers the question.
